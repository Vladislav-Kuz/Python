Создать новый столбец в таблице с пингвинами, который будет отвечать  за показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35) **Текст, выделенный полужирным шрифтом**

```python
penguins.loc[penguins['bill_length_mm']>=42, 'high_group'] = 'high'
penguins.loc[(penguins['bill_length_mm']<42) & (penguins['bill_length_mm']>=35), 'high_group'] = 'medium'
penguins.loc[penguins['bill_length_mm']<35, 'high_group'] = 'low'

penguins.sample(50)
```

**Изобразить гистограмму по flipper_length_mm с оттенком height_group. Сделать анализ**

```python
sns.histplot(penguins, x='flipper_length_mm', hue ='high_group')
```
