>**В ячейке ниже представлен код генерирующий DataFrame, которая состоит всего из 1 столбца. Ваша задача перевести его в one hot вид. Сможете ли вы это сделать без get_dummies?**





```python
import pandas as pd
import random
```

```python
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
#data.head()
data.sample(10)
```

```python
dt = data #Скопируем, чтобы не портить DataFrame
```

```python
dt.loc[dt['whoAmI'] == 'robot', 'robot'] = '1'
dt.loc[dt['whoAmI'] == 'human', 'human'] = '1'
dt = dt.fillna(0)
dt.drop('whoAmI', axis= 1 , inplace= True )
dt.sample(20)
```
